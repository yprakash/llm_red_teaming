```python
import logging
import os
import sys
import time

from dotenv import load_dotenv

load_dotenv()
print(f"Loaded dotenv: {'HF_TOKEN' in os.environ.keys()}")
```

```python
import torch

from transformers import (AutoModelForCausalLM, AutoTokenizer,
    pipeline, TextGenerationPipeline)
```

```python
# Constants
# model_id = "sarvamai/sarvam-1"
model_id = "mlx-community/sarvam-1-4bit"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_id)
print(tokenizer.__class__.__name__, "-", tokenizer.name_or_path)
print(type(tokenizer), tokenizer.vocab_size)
print(tokenizer.decode(tokenizer("hello world")["input_ids"]))
```

```python
tokenizer.pad_token_id = tokenizer.eos_token_id
tokenizer.pad_token_id
```

```python
tt = int(1000 * time.time())
model = AutoModelForCausalLM.from_pretrained(model_id, dtype="auto", device_map="auto")
print(f"Time taken to load model {int(1000 * time.time()) - tt} ms")
print(type(model))
```

```python
pipe = TextGenerationPipeline(
    model=model,
    tokenizer=tokenizer,
    device="auto",
    torch_dtype="auto",
    return_full_text=False
)
print(type(pipe))
print(pipe("hi"))
```
